```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Downloads\10th_board_results.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates(subset=['Roll_Number'], inplace=True)
```

```python
top_scorers = df[df['Percentage'] > 90]
print(top_scorers.head())
```

```python
grade_counts = df['Grade'].value_counts()
print(grade_counts)
```

```python
low_attendance = df[df['Attendance (%)'] < 75]
print(low_attendance.head())
```

```python
#NumPy
```

```python
total_marks_array = df['Total_Marks'].values
average_marks = np.mean(total_marks_array)
print("Average Total Marks:", average_marks)
```

```python
print("Max Marks:", np.max(total_marks_array))
print("Min Marks:", np.min(total_marks_array))
```

```python
std_marks = np.std(total_marks_array)
print("Standard Deviation of Total Marks:", std_marks)
```

```python
percent_array = df['Percentage'].values
above_85 = percent_array[percent_array > 85]
print(above_85)
```

```python
#MatPlot
```

```python
avg_marks_city = df.groupby('City')['Total_Marks'].mean()
avg_marks_city.plot(kind='bar', color='skyblue', title='Average Marks by City', figsize=(10,6))
plt.ylabel('Average Marks')
plt.show()
```

```python
result_status_counts = df['Result_Status'].value_counts()
result_status_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'], title='Result Status Distribution')
plt.show()
```

```python
sns.boxplot(x='Grade', y='Total_Marks',data=df)
plt.title('Marks Distribution by Grade')
plt.show()
```

```python
sns.countplot(x='Extracurricular_Activities', data=df, palette='Set2')
plt.title('Extracurricular Activities Participation')
plt.show()
```

```python
city_counts = df['City'].value_counts()
```

```python
City_percent = city_counts / city_counts.sum() * 100
```

```python
City_percent
```

```python
plt.figure(figsize=(8,8))
City_percent.plot.pie(autopct='%1.1f%%', startangle=140, cmap='Set3',legend=True)
plt.title('Percentage of Students by City')
plt.ylabel('')
plt.legend(loc='center left', bbox_to_anchor=(1, 1))
plt.show()
```
